<h1><center>Inverse filtering - Example</center></h1>

#### Import modules

```python
import IPython.display as ipd
import soundfile as sf
import librosa
import sys  
sys.path.insert(0, '../')
from src.utils import fft_plot, wave_plot, load_audio_file
from src.filters import inverse_filter
```

#### Load audio

```python
# Path
file_path = "../data/rec1_2_1n.wav"

# Load audio
s, sr = load_audio_file(file_path)
```

#### Before inverse filtering

```python
# Plot waveform
wave_plot(s, sr, "Neutral Speech")
```

```python
# Play audio sample
ipd.Audio(file_path)
```

```python
# Plot FFT
fft_plot(s, sr, "Neutral Speech - FFT")
```

#### After inverse filtering

```python
s_filtered = inverse_filter(s, 10)
```

```python
wave_plot(s_filtered, sr, "Filtered Speech")
```

```python
sf.write('../data/filtered.wav', s_filtered, sr)
```

```python
ipd.Audio('../data/filtered.wav')
```

```python
fft_plot(s_filtered, sr, "Filtered Speech - FFT")
```
